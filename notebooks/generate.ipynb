```python
from datetime import datetime, timedelta

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
start = datetime(2020, 1, 1)
stop = datetime(2020, 1, 2)
interval = timedelta(hours=1)

time = start
index = []
while time <= stop:
    index.append(time)
    time += interval
```

```python
df = pd.DataFrame(
    index=index, 
    data=np.zeros((25,5)), 
    columns=["Open", "High", "Low", "Close","Volume"],
)
```

```python
df["Open"] = df["Close"] = 2 * np.cos(np.linspace(0, 2 * np.pi, 25)) + 10
df["High"] = df["Open"] + .1
df["Low"] = df["Open"] - .1
df["Volume"] = 12
df.plot()
df.to_csv("/Users/james/Repos/My Ideas/trading-engine/framework/tests/data/BTC-BNB-1:00:00.csv", encoding="utf-8")
```
